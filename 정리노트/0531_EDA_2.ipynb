### 양성, 음성 검사자 구분하기

```python
# 데이터 타입들 중 칼럼들이 서로 상관이 있거나, 반대 성향(경향)이 있으면 서로 비교 -> 상관관계, 상관계수
pos = kr['확진자']
neg = kr['결과 음성']
```

### 사망율(치명율) 구하기

```python
# 두 데이터(컬럼) 합쳐서 하나의 결과를 도출 -> 컬럼 선택
# 확진자 수 중 사망한 사람의 숫자를 확인하면 ->  사망률을 알 수 있다.
# 확진자 수 : pos
pos
death = kr['사망']
# death/pos*100 -> 비율
```

```python
d_p = round((death / pos) * 100, 2)
d_p.head() # 치명률
```

### 완치율 구하기

```python
# 완치율
# 확진자 수
# 격리해제 수
rec = kr['격리해제']

# rec/pos*100-> 비율
```

```python
r_p = round((rec / pos) * 100, 2)
r_p.head() # 완치율
```

```python
# 치명률, 완치율 -> 그래프로 그려봤을 때 완치율이 눈에 띄게(유의미하게) 증가

plt.xticks(rotation=90)
plt.plot(d_p)
plt.plot(r_p, 'r')
plt.show()
```
