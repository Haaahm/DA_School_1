### 이상치 확인하기

- 이상치(outlier): 사분위에서 특정 범위(+-IQR*1.5, 상위경계 & 하위 경계)를 초과하는 경우

image.png

```python
import matplotlib.pyplot as plt
```

```python
# 이상치 확인 -> boxplot

plt.boxplot(out_df)
plt.show()
```

```python
out_df['국어'] = out_df['국어'].replace(810, 81) # 이상치 수정
```

```python
plt.boxplot(out_df)
plt.show()
```

```python
# 1반 학생들의 국어(수학, 영어) 성적 평균
out_df.mean()
```

```python
# bar 그래프 bar(x,y)

plt.bar(out_df.columns, out_df.mean())
plt.show()
```

```python
# 과목별로 최대 최소값
out_df.max()
```

```python
out_df.min()
```

```python
# 순서대로 -> 사분위값 -> 상위 몇 %

q1 = out_df.quantile(0.25)
q1
```

```python
q2 = out_df.quantile(0.5)
q2
```

```python
q3 = out_df.quantile(0.75)
q3
```

```python
IQR = q3 - q1
IQR             # -> 50%에 해당되는 데이터

# -(1.5*IQR) < 데이터범위 < +(1.5*IQR)   == 일반적인 데이터 허용범위
```

```python
# 국어 점수 이상치 판별 함수
def kor_outlier(df):
    kor_score = df['국어']
    if kor_score > q3['국어'] + 1.5 * IQR['국어'] or kor_score < q1['국어'] - 1.5 * IQR['국어']:
        return True
    return False

out_df['국어_이상치'] = out_df.apply(kor_outlier, axis=1)
out_df
```

### groupby를 이용해 월별 데이터 그래프 그려보기

```python
# groupby : 기준 데이터 -> (처리할)수치 데이터
# 월 별 / 기온, 풍속 ...
```

```python
weather_df.index
```

```python
weather_date = pd.Series(weather_df.index)
weather_date
```

```python
weather_date.dt.month
```

```python
weather_df['월'] = weather_date.dt.month.values
```

```python
weather_df
```

```python
# groupby
month_low_temp = weather_df.groupby(weather_df['월'])['최저기온(°C)'].mean()
```

```python
month_high_temp = weather_df.groupby(weather_df['월'])['최고기온(°C)'].mean()
```

```python
month_daily_vol = weather_df.groupby(weather_df['월'])['일강수량(mm)'].mean()
```

```python
month_wind_speed = weather_df.groupby(weather_df['월'])['최대 순간 풍속(m/s)'].max()
```

```python
month_wind_mean = weather_df.groupby(weather_df['월'])['평균 풍속(m/s)'].mean()
```

```python
month_wind_dic = weather_df.groupby(weather_df['월'])['최대 순간 풍속 풍향(hhmi)'].mean()
```

```python
x_month = weather_df['월'].unique()
```

```python
# 그룹바이 월 별 데이터 -> 1월 데이터들 -> 그래프
plt.figure(figsize=(10, 10))

# 1
plt.subplot(2, 3, 1)
plt.title('최저기온')
plt.ylabel('기온')
plt.xlabel('월')
plt.plot(x_month, month_low_temp)

# 2
plt.subplot(2, 3, 2)
plt.title('최고기온')
plt.ylabel('기온')
plt.xlabel('월')
plt.plot(x_month, month_high_temp)

# 3
plt.subplot(2, 3, 3)
plt.title('월별 일 강수량')
plt.ylabel('강수량')
plt.xlabel('월')
plt.plot(x_month, month_daily_vol) 

# 4
plt.subplot(2, 3, 4)
plt.title('월별 최대 순간 풍속')
plt.ylabel('풍속')
plt.xlabel('월')
plt.plot(x_month, month_wind_speed)

# 5
plt.subplot(2, 3, 5)
plt.title('월별 평균 풍속')
plt.ylabel('풍속')
plt.xlabel('월')
plt.plot(x_month, month_wind_mean)

# 6
plt.subplot(2, 3, 6)
plt.title('최대 순간 풍속 풍향')
plt.ylabel('풍향')
plt.xlabel('월')
plt.plot(x_month, month_wind_dic) 

plt.show()
```
