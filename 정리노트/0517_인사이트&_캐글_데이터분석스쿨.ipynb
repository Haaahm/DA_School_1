## 내적인 변수 및 자료(기획)

* 가장 먼저 원하는 주택의 가격대를 설정할 수 있도록하고,

* 주거 유형(아파트/빌라/단독/오피스텔/원룸 등)을 나누고
 
* 건물 관련 내용(건축 연도, 전용면적,주차장 1가구당 주차가능 대수, 화장실 수, 난방 유형, 풀옵션, 남향, 조망권, 브랜드),
 
* 거래 유형(매매/전세/월세 등)을 나눠서 분석을 해야한다고 봅니다.
 
* 전세의 경우엔 최근 전세사기 관련 이슈가 크니 사기에 연루될 가능성을 보여줄수 있는 지표도 보여줄 수 있으면 좋겠다는 생각이 듭니다.(내집스캔 앱)
 
* 또한 집값에 영향을 줄 수 있는 변수
 
* (주거지역인지 공업지역인지 업무 또는 상업지역 여부, 그린벨트, 교통, 교육, 관련 정책, 개발 이슈, 시장상황, 리모델링)등을 함께 
 
* 인터랙티브하게 볼 수 있으면 좋을 듯 싶습니다.
 
* 회귀 시계열 분석으로 예측 진행하고 ML 등으로 오차(이상치, 결측값)를 보정할 수 있지 않을까 싶습니다.
 (세세한 분석 방법에 대해서는 조금 더 고민이 필요한 부분)
 
* 고객이 원하는 변수들을 설정하고 컴팩트한 선그래프와 함께 설명을 곁들이면 좋지 않을까 생각이 듭니다.
 
* 그리고 이 집을 구매하는 사람들의 유형 같은 것도 같이 분석하여 표시하면 구매자들에게 도움이 될 것 같습니다.

## 부동산 문제점 (외적인 부분)
1. 갭 투자의 증가
갭 투자는 매매 가격과 전세 가격의 차액만큼만 투자하여 집을 산 후 집 값이 오르면 투자 수익을 실현하는 투자 기법. 구매자의 지분보다 세입자의 지분이 더 큰 경우가 대부분이므로, 집 값 하락으로 인한 피해를 세입자가 더 크게 받게 된다.

2. 부동산 임대사업자 우대 정책
정부에서 임대사업자를 장려하면서, 전국적으로 임대 주택이 크게 증가하였고, 적은 자본으로 많은 임대 주택을 소유한 사업자들이 늘었다. 이런 상황에서 주택 시작이 조금이라도 충격을 받게 되면, 많은 세입자들이 피해를 받게 되는 환경이 조성.

3. 전세 사기
    의도적으로 허위 매물이나 불량 매물을 이용하여 세입자를 속이는 경우.


3. 깡통 전세의 대상
    1. 신축 빌라
    2. 단독 주택
    3. 전세가율이 높고, 매물이 적은 지역


목표 서비스
1. 본 프로젝트는 주택의 적절한 시장가격과 그 가격의 변동을 예측하여, 정보 격차로 인한 깡통 전세의 피해를 예방하자는 목표를 정하고,
2. 그 대상을 피해 대상이 되기 쉬운 연립/다세대 주택, 오피스텔 로 한다.

결과적으로 완성된 모델을 통하여
1. 집의 조건과 비교할 전세 보증금 액수가 주어진다면,
2. 집의 매매 가격을 예측할 수 있고,
3. 예상 가격과 깡통 전세 기준 비율을 바탕으로 보증금이 과한 수준인지 직접적으로 비교하여,
4. 위험 정도를 판단하는 기준을 제공할 수 있다.
